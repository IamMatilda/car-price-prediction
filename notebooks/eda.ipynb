```python
import pandas as pd

df = pd.read_csv("C:\\Users\\matve\\OneDrive\\Рабочий стол\\CAR-PRICE-PREDICTION-WITH-MACHINE-LEARNING-main\\Data\\used_car_dataset.csv")

print(df.shape)
df.head()
```

```python
features = [
    "Brand",
    "model",
    "Year",
    "kmDriven",
    "Transmission",
    "Owner",
    "FuelType"
]

target = "AskPrice"
```

```python
print(df.dtypes)
```

```python
df["kmDriven"] = (
    df["kmDriven"]
    .str.replace(" km", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["kmDriven"] = pd.to_numeric(df["kmDriven"], errors="coerce")

df["AskPrice"] = (
    df["AskPrice"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

df["AskPrice"] = pd.to_numeric(df["AskPrice"], errors="coerce")
```

```python
print(df[["kmDriven", "AskPrice"]].head())
print(df[["kmDriven", "AskPrice"]].dtypes)
```

```python
print(df.isna().sum().sort_values(ascending=False))
```

```python
df = df.dropna(subset=["kmDriven"])
```

```python
print(df.shape)
print(df.isna().sum())
```

```python
df = df.drop(
    columns=[
        "Age",
        "PostedDate",
        "AdditionInfo"
    ]
)
```
